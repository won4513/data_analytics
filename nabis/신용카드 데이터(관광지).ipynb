```python
popu=pd.read_csv("C:/Users/guswn/OneDrive/바탕 화면/생활인구/df.csv",encoding="utf-8")
```

```python
pop=popu[["행정구역","2022평균"]]
```

```python
pop["2022평균"] = pop["2022평균"].apply(lambda x: float(str(x).replace(',', '')))
```

```python
columns_to_process = new.columns[4:5]  # 첫 번째 열,두번째열

for column in columns_to_process:
    new[column] = new[column].apply(lambda x: int(str(x).replace(',', '')))
```

```python
df_grouped = new.groupby([ 'sgg_cd', '시군구명', '대분류명']).sum().reset_index()
```

```python
df=df_grouped.drop(["sgg_cd","기준년월"],axis=1)
```

```python
merged_df = pd.merge(df, pop, left_on='시군구명', right_on='행정구역', how='left')
```

```python
merged_df["생활인구별 매출금액"]=merged_df["추정매출금액"]/merged_df["2022평균"]
```

```python
merged_df["생활인구별 매출건수"]=merged_df["추정매출건수"]/merged_df["2022평균"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='Malgun Gothic')
sns.set(font="Malgun Gothic",#"NanumGothicCoding", 
        rc={"axes.unicode_minus":False}, # 마이너스 부호 깨짐 현상 해결
        style='darkgrid')
# 데이터를 데이터프레임으로 변환합니다. (주어진 데이터를 변수 df에 저장한다고 가정합니다.)
# 위의 데이터를 사용하여 데이터프레임을 생성합니다.

# 시군구와 대분류별 추정 매출금액을 시각화합니다.
plt.figure(figsize=(14, 8))
sns.barplot(x="시군구명", y="생활인구별 매출금액", hue="대분류명", data=merged_df, ci=None)
plt.title("생활인구별 매출금액에 따른 시군구와 대분류")
plt.xlabel("시군구명")
plt.ylabel("생활인구별 매출금액")
plt.xticks(rotation=45)
plt.show()

```
